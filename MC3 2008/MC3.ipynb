```python
startdate = '2006-06-01'
enddate  = '2006-06-08'
slicedDF = FamilyDF.loc[(FamilyDF['Datetime']<=enddate) & (FamilyDF['Datetime']>=startdate)]
print (slicedDF)
```

```python
##scatter plot of interest
df2 =  slicedDF.groupby(['From','To']).size().reset_index(name='freq')
print (df2.head())

x1 = df2['From']
y1 = df2['To']
actual_data = ColumnDataSource(data=dict(x=x1,
                                             y=y1,size=df2['freq'],alpha=df2['freq']/5))
a=figure(plot_height=500, plot_width=700, title='Clustering cell tower data')
a.circle('x','y',fill_color='firebrick',line_color='yellow',source=actual_data,alpha='alpha',
         line_alpha=0.5,size='size')
output_notebook()
show(a)

```

```python
graphDF=slicedDF
N = 400
mult = 4
node_indices = list(set(set(list(FamilyDF['From'])) | set(list(FamilyDF['To']))))

plot = figure(title="Graph Layout Demonstration", x_range=(-5.1,5.1), y_range=(-5.1,5.1),
              plot_height=450,plot_width=800)
edge_hover_tool = HoverTool( tooltips=[('Calls',"@start -> @end")])
plot.add_tools(edge_hover_tool, TapTool(), BoxSelectTool())

graph_renderer = GraphRenderer()

graph_renderer.node_renderer.glyph = Circle(size=5, fill_color=Spectral4[0])
graph_renderer.node_renderer.selection_glyph = Circle(size=15, fill_color=Spectral4[2])
graph_renderer.node_renderer.hover_glyph = Circle(size=15, fill_color=Spectral4[1])

graph_renderer.edge_renderer.glyph = MultiLine(line_color="#CCCCCC", line_alpha=0.8, line_width=0.5)
graph_renderer.edge_renderer.selection_glyph = MultiLine(line_color=Spectral4[2], line_width=5)
graph_renderer.edge_renderer.hover_glyph = MultiLine(line_color=Spectral4[1], line_width=5)

graph_renderer.selection_policy = NodesAndLinkedEdges()
graph_renderer.inspection_policy = EdgesAndLinkedNodes()


graph_renderer.node_renderer.data_source.data = dict(
    index=node_indices)
graph_renderer.edge_renderer.data_source.data = dict(
    start=graphDF['From'],
    end=graphDF['To'])



### start of layout code
circ = [i*2*math.pi/N for i in node_indices]
x = [mult*math.cos(i) for i in circ]
y = [mult*math.sin(i) for i in circ]

graph_layout = dict(zip(node_indices, zip(x, y)))
graph_renderer.layout_provider = StaticLayoutProvider(graph_layout=graph_layout)

plot.renderers.append(graph_renderer)

output_file("graph.html")
show(plot)
```

```python
#####################################################################################################
from bokeh.layouts import row, layout
datelist =['2006-06-01','2006-06-02','2006-06-03','2006-06-04','2006-06-05',
           '2006-06-06','2006-06-07','2006-06-08','2006-06-09','2006-06-10']

#widget
startdate_select = Select(value='2006-06-01',
                          title='Select start date',
                          width=200,
                          options=datelist)
enddate_select = Select(value='2006-06-02',
                          title='Select end date',
                          width=200,
                          options=datelist)



#callback
def update_startdate_attribute(attrname,old,new):
    startdate = startdate_select.value
    enddate = enddate_select.value
    slicedDF = FamilyDF.loc[(FamilyDF['Datetime']<=enddate) & (FamilyDF['Datetime']>=startdate)]
    #scatter
    df2 =  slicedDF.groupby(['From','To']).size().reset_index(name='freq')
    x1 = df2['From']
    y1 = df2['To']
    actual_data.data = dict(x=x1,y=y1,size=5*df2['freq'],alpha=df2['freq']/5)
    #graph
    graphDF=slicedDF
    print (graphDF)
    node_indices = list(set(set(list(graphDF['From'])) | set(list(graphDF['To']))))
    graph_renderer.node_renderer.data_source.data = dict(index=node_indices)
    graph_renderer.edge_renderer.data_source.data = dict(start=graphDF['From'],end=graphDF['To'])
    print ('in the func')
    circ = [i*2*math.pi/N for i in node_indices]
    circ = [i*2*math.pi/N for i in node_indices]
    x = [mult*math.cos(i) for i in circ]
    y = [mult*math.sin(i) for i in circ]
    graph_layout = dict(zip(node_indices, zip(x, y)))
    graph_renderer.layout_provider = StaticLayoutProvider(graph_layout=graph_layout)
    plot.renderers.append(graph_renderer)
    print ('a')
    print (startdate,enddate)


    
def update_enddate_attribute(attrname,old,new):
    startdate = startdate_select.value
    enddate = enddate_select.value
    slicedDF = FamilyDF.loc[(FamilyDF['Datetime']<=enddate) & (FamilyDF['Datetime']>=startdate)]
    #scatter
    df2 =  slicedDF.groupby(['From','To']).size().reset_index(name='freq')
    x1 = df2['From']
    y1 = df2['To']
    actual_data.data = dict(x=x1,y=y1,size=5*df2['freq'],alpha=df2['freq']/5)
    #graph
    graphDF=slicedDF
    print (graphDF)
    node_indices = list(set(set(list(graphDF['From'])) | set(list(graphDF['To']))))
    graph_renderer.node_renderer.data_source.data = dict(index=node_indices)
    graph_renderer.edge_renderer.data_source.data = dict(start=graphDF['From'],end=graphDF['To'])
    print ('in the func')
    circ = [i*2 for i in node_indices]
    circ = [i*2*math.pi/N for i in node_indices]
    x = [mult*math.cos(i) for i in circ]
    y = [mult*math.sin(i) for i in circ]
    graph_layout = dict(zip(node_indices, zip(x, y)))
    graph_renderer.layout_provider = StaticLayoutProvider(graph_layout=graph_layout)
    plot.renderers.append(graph_renderer)
    print ('b')
    print (startdate,enddate)

    

startdate_select.on_change('value', update_startdate_attribute)
enddate_select.on_change('value', update_enddate_attribute)
```

```python
l=layout ([
    [row([startdate_select,enddate_select]) ],
    [plot],
       [a]
])
curdoc().add_root(l)
show(l)
```
