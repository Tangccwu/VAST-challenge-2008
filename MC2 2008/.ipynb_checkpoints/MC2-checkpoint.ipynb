```python
import xml.etree.ElementTree as ET
import pandas as pd

xml_data = open('GEOSPATIALTEMPORAL/Migrant Data.xml').read()

def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
            all_records.append(record)
    return pd.DataFrame(all_records)
```

```python
df = xml2df(xml_data)
print (df.head())

locations=[]
for row in df.iterrows():
    if (row[1]['EncounterCoords'] not in locations):
        locations.append(row[1]['EncounterCoords'])
    if (row[1]['LaunchCoords'] not in locations):
        locations.append(row[1]['LaunchCoords'])
print (locations[1])

locationDF = pd.DataFrame(columns=['Node','xcoord','ycoord'])
for i in range(0,len(locations)):
    if(locations[i]!=None):
        location = locations[i].split(',')
        locationDF.loc[len(locationDF)] = [i,location[0],location[1]]
    else:
        locationDF.loc[len(locationDF)] = [i,'-79.70','23.55']

locationDF[['xcoord','ycoord']] = locationDF[['xcoord','ycoord']].apply(pd.to_numeric)
print (len(locationDF))
```

```python
sLength=1
list1=[]
list2=[]
list3=[]
list4=[]
list5=[]
list6=[]
for row in df.iterrows():
    Encounterx = row[1]['EncounterCoords'].split(',')[0]
    Encountery = row[1]['EncounterCoords'].split(',')[1]
    if(row[1]['LaunchCoords']!=None):
        Launchx = row[1]['LaunchCoords'].split(',')[0]
        Launchy = row[1]['LaunchCoords'].split(',')[1]
    else:
        Launchx = -79.70
        Launchy = 23.55
    list1.append(Encounterx)
    list2.append(Encountery)
    list3.append(Launchx)
    list4.append(Launchy)
    list5.append(locations.index(row[1]['EncounterCoords']))
    list6.append(locations.index(row[1]['LaunchCoords']))

#print (df.head())
df['Encounterx'] = list1
df['Encountery']=list2
df['Launchx']=list3
df['Launchy']=list4
df['EncounterNode']=list5
df['LaunchNode']=list6
#df['EncounterDate'] = pd.to_datetime(df['EncounterDate'])
df[['EncounterNode','LaunchNode']] = df[['EncounterNode','LaunchNode']].apply(pd.to_numeric)
df[['Encounterx','Encountery']] = df[['Encounterx','Encountery']].apply(pd.to_numeric)
df[['Launchx','Launchy']] = df[['Launchx','Launchy']].apply(pd.to_numeric)
df[['Passengers','NumDeaths']] = df[['Passengers','NumDeaths']].apply(pd.to_numeric)
print(df.head())
```

```python
startdate = '2005-04-24'
enddate = '2005-12-31'
#graphDF = df
graphDF = df.loc[(df['EncounterDate']<=enddate) & (df['EncounterDate']>=startdate)]
```

```python

node_indices = list(range(0,len(locations)))
plot = figure(title="Migration", 
              x_range=(-82.15,-79.2), y_range=(23.35,26.15),
              plot_height=450,plot_width=800)
edge_hover_tool = HoverTool( tooltips=[('Intercepts',"@startcoord -> @endcoord"),
                                       ('Migrants','@passengers')])
plot.add_tools(edge_hover_tool, TapTool(), BoxSelectTool())

graph_renderer = GraphRenderer()

graph_renderer.node_renderer.glyph = Circle(size=5, fill_color='yellow')
graph_renderer.node_renderer.selection_glyph = Circle(size=15, fill_color=Spectral4[2])
graph_renderer.node_renderer.hover_glyph = Circle(size=15, fill_color=Spectral4[1])

graph_renderer.edge_renderer.glyph = MultiLine(line_color="#CCCCCC", line_alpha=1, line_width=0.1)
graph_renderer.edge_renderer.selection_glyph = MultiLine(line_color=Spectral4[2], line_width=5)
graph_renderer.edge_renderer.hover_glyph = MultiLine(line_color=Spectral4[1], line_width=5)

graph_renderer.selection_policy = NodesAndLinkedEdges()
graph_renderer.inspection_policy = EdgesAndLinkedNodes()


graph_renderer.node_renderer.data_source.data = dict(
    index=node_indices)
graph_renderer.edge_renderer.data_source.data = dict(
    start=graphDF['LaunchNode'],
    end=graphDF['EncounterNode'],
    startcoord = graphDF['LaunchCoords'],
    endcoord = graphDF['EncounterCoords'],
    passengers = graphDF['Passengers']
)



### start of layout code
x = locationDF['xcoord']
y = locationDF['ycoord']
#plot.patches('x','y',source=source,fill_color='yellow')
graph_layout = dict(zip(node_indices, zip(x, y)))
graph_renderer.layout_provider = StaticLayoutProvider(graph_layout=graph_layout)
plot.renderers.append(graph_renderer)

output_file("graph.html")
show(plot)
```
